## Introducción

### 1. Problemas de optimización sin restricciones

\begin{equation}
\begin{array}{llll}
\displaystyle \min_{\mathbf{x}} f(\mathbf{x}) \\
\mathbf{x} \in \mathcal{R}^n \\
\mathbf{x} = [x_1, \ldots, x_n]^{\top}\\ 
x_i \in \Re
\end{array}
\end{equation}

$x_i:$ Variables de optimización

$f: \Re^n \to \Re$ Función de costo

### 2. Problemas de optimización con restriciones

\begin{equation}
\begin{array}{lrclcl}
\displaystyle \min_{\mathbf{x}} & f(\mathbf{x}) & \\
\textrm{s.t.} & g_i(\mathbf{x}) = 0 & i = 1,\ldots,m \\
              & h_j(\mathbf{x}) \leq 0 & j =1, \ldots l
\end{array}
\end{equation}

$g_i, h_i: \Re^n \to \Re$ Restricciones


### Ejemplo 1: Regresión paramétrica
![](https://drive.google.com/uc?export=view&id=1aw7etEnC9sDJZWqNwD5-MKogWCeUaf4G)

Dado un conjunto de pares $\{x^{(i)},y^{(i)}\}_{i=1}^M$, se desea aproximar los valores de $y^{(i)}$ por medio de la siguiente función paramétrica:

\begin{equation}
\hat{y}^{(i)} = f(x^{(i)},\mathbf{\omega}) = \omega_0 + \omega_1 x^{(i)} + \omega_2 {x^{(i)}}^2
\end{equation}

Para obtener los valores de $\omega$ que mejor aproximan la función, se plantea la siguiente función de costo:

\begin{equation}
L(\mathbf{\omega}) = \sum_{i=1}^{M}(f(x^{(i)},\mathbf{\omega}) - y{(i)})^2, 
\end{equation}

que se minimiza con respecto a los parámetros desconocidos $\mathbf{\omega}$

\begin{equation}
\displaystyle \mathbf{\omega}^* = [\omega_0^*, \omega_1^*, \omega_2^*] =  \text{argmin}_{\omega} L(\omega)
\end{equation}

### Ejemplo 2: Regresión lineal

![](https://drive.google.com/uc?export=view&id=1SaUgjSaAAcfB_57KFJc1gy3XKDE5Q6lC)


La definición anterior se puede extender de forma lineal cuando $\mathbf{x}^{(i)} = [x_1^{(i)}, \ldots x_n^{(i)}]^{\top}$. En este caso, se puede plantear la función de aproximación de los valores de $y^{(i)}$ de la siguiente forma:

\begin{equation}
\hat{y}^{(i)} = f(\mathbf{x}^{(i)},\mathbf{\omega}) = \omega_0 + \omega_1 x_1^{(i)} + \ldots + \omega_n x_n^{(i)} 
\end{equation}

### Ejemplo 4: Optimización con restricciones - Asignación de recursos

| $M$ estaciones de energía        | N clientes                  |
|----------------------------------|-----------------------------|
| $S_M$ Máxima potencia estación M | $C_N$ Demanda del cliente N |

$P_{mn}:$ Costo de transporte de unidad de electricidad de la estación $M$ al cliente $N$.

Problema: Hallar $x_{mn}$ que es la cantidad de elecricidad que se debe transmitir de la estación $m$ al cliente $n$.

\begin{equation}
\begin{array}{lrclcl}
\displaystyle \min_{\mathbf{x}} & \sum_{m,n} P_{mn}x_{mn} & \\
\textrm{s.t.} & \sum_{n=1}^N x_{mn} \leq S_M, & m=1,\ldots,M \\
              & \sum_{m=1}^M x_{mn} = C_N, & n=1,\ldots,N \\
              & x_{mn} \geq 0
\end{array}
\end{equation}

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```
f = lambda x1,x2: np.cos(3*x1 + 2*x2) + np.cos(2*x1 + 4*x2) - 2*np.sin(x1 + x2)

X,Y = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
Z = f(X,Y)

fig = plt.figure(figsize=(7.2, 6.0))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z,cmap='jet',alpha=0.7)
ax.grid(False)

fig.savefig('function.png',transparent=True)
```
