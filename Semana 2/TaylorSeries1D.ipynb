```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-2,2,100)
f = 1

x = lambda t,f: np.cos(2*np.pi*f*t)
plt.plot(t,x(t,f))
```

```python
q = lambda t,f: 1 -  (2 * (np.pi**2) * (f**2) * (t**2))
plt.plot(t,q(t,f))
```

```python
t1 = np.linspace(-0.5,0.5,100)

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(t1,x(t1,f))

ax1.plot(t1,q(t1,f),'g')
```
